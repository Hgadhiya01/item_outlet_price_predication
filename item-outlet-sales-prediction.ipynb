### import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### data processing (EDA process)

```python
df = pd.read_csv(r"../input/item-outlet-sales-prediction/train.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(["int64","float64"]).keys()
num_cols
```

```python
cat_cols = df.select_dtypes(["object"]).keys()
cat_cols
```

```python
for var in num_cols:
    df[var].fillna(df[var].mean(),inplace = True)
```

```python
for var in cat_cols:
    df[var].fillna(df[var].mode()[0],inplace = True)
```

```python
df.isnull().sum()
```

### one hot encoding

```python
from sklearn.preprocessing import LabelEncoder
lr = LabelEncoder()
for var in cat_cols:
    lr.fit(df[var])
    df[var]= lr.transform(df[var])
```

### data spliting

```python
X = df.drop("Item_MRP",axis=1)
y = df["Item_MRP"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train , y_test = train_test_split(X,y, test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
lr.score(X_test,y_test)
```

```python
from sklearn.svm import SVR
svr = SVR() 
svr.fit(X_train,y_train)
y_pred = svr.predict(X_test)
svr.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train)
y_pred = dtr.predict(X_test)
dtr.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
y_pred = rfr.predict(X_test)
rfr.score(X_test,y_test)
```
